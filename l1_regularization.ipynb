```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fat matrix
N = 50
D = 50
```

```python
# X is uniformly distributed points of size NxD, centered around 0 from -5 to +5
X = (np.random.random((N,D)) - 0.5) * 10
X
```

```python
true_w = np.array([1, 0.5, -0.5] + [0]*(D-3))
true_w
```

```python
# X dot true_w plus some gaussian random noise which is the assumption in linear regression
Y = X.dot(true_w) + np.random.randn(N)*0.5
Y
```

```python
costs = []
w = np.random.randn(D) / np.sqrt(D)
learning_rate = 0.001
l1 = 10.0
for t in range(500):
    Yhat = X.dot(w)
    delta = Yhat - Y
    w = w - learning_rate*(X.T.dot(delta) + l1*np.sign(w))
    mse = delta.dot(delta) / N
    costs.append(mse)
```

```python
costs
```

```python
plt.plot(costs)
plt.show()
```

```python
print('Final w:', w)
```

```python
plt.plot(true_w, label='true w')
plt.plot(w, label='w_map')
plt.legend()
plt.show()
```
