```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load data
X = []
Y = []
for line in open('data_1d.csv'):
    x, y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
```

```python
# let's turn X and Y into numpy arrays
X = np.array(X)
Y = np.array(Y)
```

```python
print(X)
print(Y)
```

```python
# plot to see what it looks like
plt.scatter(X, Y)
plt.show()
```

```python
# apply the equations that we learned to calculate a and b
denominator = X.dot(X) - X.mean()*X.sum()
a = ( X.dot(Y) - Y.mean()*X.sum() ) / denominator
b = ( Y.mean()*X.dot(X) - X.mean()*X.dot(Y) ) / denominator
```

```python
print(denominator)
print('Slope: %f' % a)
print('Y-intercept: %f' % b)
```

```python
# calculate predicted Y
Yhat = a*X + b
```

```python
print(Yhat)
```

```python
# plot it all
plt.scatter(X, Y)
plt.plot(X, Yhat)
plt.show()
```

```python
# calculate r-squared
d1 = Y - Yhat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1) / d2.dot(d2)
print('r-squared: %f' % r2)
```
